### Sistematično iskanje grafov z maksimalno sigma čisto nepravilnostjo na majhnih grafih

```python
# generate all triangle free graphs with n vertices
def generate_triangle_free_graphs(n):
    return [G for G in graphs.nauty_geng(f"{n} -t")]
```

```python
from itertools import combinations
# sigma total irr
def sigma_irr(G):
    return sum((u - v) ** 2 for u, v in combinations(G.degree(), 2))
```

```python
# checking for all small n
for n in range(1,13):
    #triangle_free_graphs = generate_triangle_free_graphs(n) # generate triangle free graphs
    triangle_free_graphs = graphs.nauty_geng(f"{n} -t")
    max_sigma = 0 # store current max sigma value
    graphs_with_max_sigma = [] # store graphs with max sigma value

    # find graphs with max sigma value
    for G in triangle_free_graphs:

        sigma = sigma_irr(G) # get sigma value of graph

        if sigma > max_sigma: # update if its the best
            graphs_with_max_sigma = [G]
            max_sigma = sigma
        
        elif sigma == max_sigma: # if equal, just append
            graphs_with_max_sigma.append(G)
    
    # show results
    print(f"Red: {n}, Max sigma: {max_sigma}")#{len(triangle_free_graphs)}")
    for i, graph in enumerate(graphs_with_max_sigma):
        graph.show(title=f"Graph {i + 1}, {max_sigma}", vertex_size=300, vertex_color="gold")
```
